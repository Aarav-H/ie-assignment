```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Gravity_V202211.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sort_df=df[['year','country_id_o','country_id_d','tradeflow_baci' ]]
sort_df.head()
```

```python
sort_df['tradeflow_baci'].isnull().sum()
```

```python
sort_df=sort_df[(sort_df['country_id_o']=='IND')]
sort_df.head()
```

```python
target_countries=['AUS','CHN','BRN','CAN','CHL','IDN.2','JPN','KOR','MYS.2','MEX','NZL','PNG','PER','PHL','RUS','SGP','THA','USA','VNM.2']
sort_df=sort_df[sort_df['country_id_d'].isin(target_countries)]
sort_df.head()
```

```python
sort_df.dtypes
```

```python
##sort_df.to_excel('output.xlsx')
```

```python
sort_df["year"]=sort_df["year"].astype(str)
##sort_df['year'].apply(str)
target_years=['2015','2016','2017','2018','2019','2020']
sort_df=sort_df[sort_df['year'].isin(target_years)]
sort_df.head()
```

```python
order=['AUS','CHN','BRN','CAN','CHL','IDN.2','JPN','KOR','MYS.2','MEX','NZL','PNG','PER','PHL','RUS','SGP','THA','USA','VNM.2']
sort_df['country_id_o'] = pd.Categorical(sort_df['country_id_o'], categories=order, ordered=True)

sort_df = sort_df.sort_values(by=['year', 'country_id_o'])

sort_df.head()
```

```python
sort_df.to_excel('output.xlsx')
```
